# The model of Yuri Evgenievich Nesterov
$I(d) = \gamma(d)H(d)$ 


$T(d) = T(d-1) + C(d)$


$C(d) = I(d-\Delta)$


$T(d+1) = T(d) + \gamma(d-\Delta+1)(T(d) - T(d-\Delta))$.


$\gamma(d) = \frac{I(d)}{H(d)} = \frac{C(d+\Delta)}{T(d+\Delta-1) - T(d-1)}$

$ \Gamma(d) = \sum_{i=0}^{\Delta-1}\gamma(d+i)$

we will consider the delay of $\Delta$=10

```python
#HIT

#H=the number of asymptomatic virus holders 
def H(d,Delta):
    return T[d+Delta-1]-T[d-1]
Delta=10
h=["_"]
for d in range(1,371-(Delta-1)):
        h.append(H(d,Delta))
for y in range(0,9):
    h.append('_') #WE NEED TO KNOW MORE INFORMATION
df.insert(3,"H", h, True) 


#gamma=daily infection rates

def gamma(d, Delta):
        hh=T[d+Delta-1]-T[d-1]
        return (T[d+Delta]-T[d+Delta-1])/hh if hh != 0 else 0
g=[1/Delta]
for d in range(1,371-(Delta)):
        g.append(gamma(d,Delta))
for r in range(0,10):
    g.append('_') #If we can't use the formula,so we consider it=1/Delta
df.insert(4,"GAMMA", g, True) 
ga=df['GAMMA']


#I=the number of newly infected persons during the day d
def I(d,Delta):
    return C[d+Delta]
Delta=10
II=[]
for d in range(0,371-10):
     II.append(I(d,Delta))  
for ii in range(0,10):
    II.append('_') #WE NEED TO KNOW MORE INFORMATION
df.insert(5,"I", II, True)

df
```

# Model with approximation of the $\gamma$ 
Within the framework of this model, it is impossible to build a chart for the current day. It is proposed to use machine learning methods to predict the $\gamma$ parameter

```python
time=date[:371-10]
g1=g[:371-10]
data1= pd.Series(g1, index=time)
data1               
```

A popular and widely used statistical method for time series forecasting is the ARIMA model.

ARIMA is an acronym that stands for AutoRegressive Integrated Moving Average. It is a class of model that captures a suite of different standard temporal structures in time series data.
The parameters of the ARIMA model are defined as follows:

p: The number of lag observations included in the model, also called the lag order.

d: The number of times that the raw observations are differenced, also called the degree of differencing.

q: The size of the moving average window, also called the order of moving average.

```python
gamma_model = ARIMA(data1, order=(6, 0, 4), trend='n').fit()
```

```python
for key in date[371-10:]:
        cur_date = key.strftime('%Y-%m-%d')
        MA = gamma_model.predict(start=cur_date, end=cur_date).values[0]
        g1.append(MA)
```
